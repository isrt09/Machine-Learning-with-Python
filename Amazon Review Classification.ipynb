```python
amazon = amazon.drop(['date','rating'],axis=1)
amazon.head()
```

```python
dummy_variation = pd.get_dummies(amazon['variation'], drop_first=True)
dummy_variation
```

```python
amazon.drop('variation', axis=1, inplace=True)
amazon.head()
```

```python
amazon = pd.concat([amazon,dummy_variation], axis=1)
amazon.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
amazon_vectorizer = vectorizer.fit_transform(amazon['verified_reviews'])
amazon_vectorizer.shape
```

```python
# print(vectorizer.get_feature_names())
print(amazon_vectorizer.toarray())
```

```python
amazon.head()
```

```python
amazon.drop('verified_reviews', axis=1, inplace=True)
amazon.head()
```

```python
encoded_reviews = pd.DataFrame(amazon_vectorizer.toarray())
encoded_reviews
```

```python
amazon = pd.concat([amazon,encoded_reviews],axis=1)
amazon
```

```python
X = amazon.drop('feedback',axis=1)
X.shape
```

```python
y = amazon['feedback']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier = RandomForestClassifier(n_estimators=100, criterion ='entropy')
randomforest_classifier.fit(X_train,y_train)
```

```python
y_predict_train = randomforest_classifier.predict(X_train)
```

```python
y_predict_train
```

```python
y_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_train, y_predict_train))
```
