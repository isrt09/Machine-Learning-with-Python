#**PROBLEM STATEMENT**<br/>
Dataset includes house sale prices for King County in USA.
Homes that are sold in the time period: May, 2014 and May, 2015.<br/>

**Data Source:** https://www.kaggle.com/harlfoxem/housesalesprediction

**Data Dictionary:**

**ida:** notation for a house

**date:**Date house was sold

**price:** Price is prediction target

**bedrooms:** Number of Bedrooms/House

**bathrooms**: Number of bathrooms/House

**sqft_living**: square footage of the home

**sqft_lot**: square footage of the lot

**floors**: Total floors (levels) in house

**waterfront**: House which has a view to a waterfront

**view**: Has been viewed

**condition**: How good the condition is ( Overall )

**grade**: overall grade given to the housing unit, based on King County grading system

**sqft_abovesquare**: footage of house apart from basement

**sqft_basement**: square footage of the basement

**yr_built**: Built Year

**yr_renovated**: Year when house was renovated

**zipcode**: zip

**lat**: Latitude coordinate

**long**: Longitude coordinate

**sqft_living15**: Living room area in 2015(implies-- some renovations)

**sqft_lot15:** lotSize area in 2015(implies-- some renovations)

```python
selected_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement']
```

```python
X= df[selected_features]
```

```python
y = df['price']
```

```python
X.shape
```

```python
y.shape
```

```python
y = y.values.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()

X_scaled = sc.fit_transform(X)
y_scaled = sc.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_scaled,test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(50, input_dim = 7, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
epochs_model = model.fit(X_train,y_train, epochs=100, batch_size=50, validation_split=0.2)
```

```python
epochs_model.history.keys()
```

```python
plt.plot(epochs_model.history['loss'])
plt.plot(epochs_model.history['val_loss'])
plt.xlabel('Epoch')
plt.ylabel('Training and Validation Loss')
plt.title('Model Loss Progress during Training')
plt.legend(['Training Loss','Validation Loss'])
plt.show()
```

```python
# ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement']
X_test_1 = np.array([[ 4, 3, 1960, 5000, 2, 2000, 3000 ]])
sc_1 = MinMaxScaler()
X_test_1_scaled = sc_1.fit_transform(X_test_1)
y_pred_scaled   = model.predict(X_test_1_scaled)
y_pred_scaled
```

```python
y_predict = model.predict(X_test)
plt.plot(y_test, y_predict, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
y_pred_orginal  = sc.inverse_transform(y_predict)
y_test_original = sc.inverse_transform(y_test)
plt.plot(y_test_original, y_pred_orginal, "^", color = 'b')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
plt.xlim(0, 5000000)
plt.ylim(0, 3000000)
```

```python
k = X_test.shape[1]
n = len(X_test)

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test_original, y_pred_orginal)),'.3f'))
MSE = mean_squared_error(y_test_original, y_pred_orginal)
MAE = mean_absolute_error(y_test_original, y_pred_orginal)
r2 = r2_score(y_test_original, y_pred_orginal)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR-squared (R²) =', r2, '\nAdjusted R² =', adj_r2) 
```
